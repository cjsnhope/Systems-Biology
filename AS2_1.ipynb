```python
# load libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# define the differential equation model
def model(state, t, I, E, a, b, c):
    X, Y, Z = state
    # same formula as written on the slide
    dXdt = I - (a * E * X)
    dYdt = (a * E * X) - (b * Y)
    dZdt = (b * Y) - (c * Z)
    return [dXdt, dYdt, dZdt]

# time setting
t = np.arange(0, 30.2, 0.2)

# define drawing graph helper function (almost same as the function in #AS1)
def plot_simulation(t, result, title):
    plt.figure(figsize=(6.1, 6))

    # set the color and width of line (x: blue, y: red, z: yellowgreen)
    plt.plot(t, result[:, 0], color='blue', label='x', linewidth=1.2)
    plt.plot(t, result[:, 1], color='red', label='y', linewidth=1.2)
    plt.plot(t, result[:, 2], color='#66FF66', label='z', linewidth=1.2)

    # label
    plt.title(title, pad=10, fontsize=10, fontweight='bold')
    plt.xlabel('Time', fontsize=9)
    plt.ylabel('Concentrations', fontsize=9)

    # set the range and ticks of x axis
    plt.xlim(-1.5, 31.5) # range including blank
    plt.xticks([0, 5, 10, 15, 20, 25, 30]) # ticks

    # set the range and ticks of y axis
    plt.ylim(-0.1, 2.1)
    plt.yticks([0.0, 0.5, 1.0, 1.5, 2.0], rotation=90, va='center') # ticks

    # set the legend
    leg = plt.legend(loc='upper right', ncol=3, edgecolor='black', framealpha=1,
                 handlelength=2, columnspacing=1.0, bbox_to_anchor=(0.98, 0.98),
                 borderpad=0.9, fancybox=False)
    for line in leg.get_lines():
      line.set_linewidth(1.0)

    # adjust a layout
    plt.tight_layout()

    # show the graph
    plt.show()
```

```python
##### Initial condition 1 #####

# set the condition as written on the slide
Input_1 = 1.0
Enzyme_1 = 1.0
state_1 = [0.0, 0.0, 0.0]  # [x, y, z] = 0mM

# parameters calculated so that the steady state converges to approximately 0.67 (2/3)
# (to make same graph on the slide)
a1, b1, c1 = 1.5, 1.5, 1.5

# get the result and draw a graph 1
result_1 = odeint(model, state_1, t, args=(Input_1, Enzyme_1, a1, b1, c1))
plot_simulation(t, result_1, 'Simulation 1: I=1, x=0, y=0, z=0')
```

```python
##### Initial condition 2 #####

# set the condition as written on the slide
Input_2 = 0.75
Enzyme_2 = 1.0
state_2 = [1.0, 1.0, 1.0]  # [x, y, z] = 1mM

# parameters as condition 1
# (to make same graph on the slide)
a2, b2, c2 = a1, b1, c1

# get the result and draw a graph 2
result_2 = odeint(model, state_2, t, args=(Input_2, Enzyme_2, a2, b2, c2))
plot_simulation(t, result_2, 'Simulation 2: I=0.75, x=1, y=1, z=1')
```

```python
##### Initial condition 3 #####

# set the condition as written on the slide
Input_3 = 0.75
Enzyme_3 = 1.0
state_3 = [1.0, 1.0, 1.0]  # [x, y, z] = 1mM

# new parameters that make x converge to 0.5 and y and z to about 1.1
# (to make same graph on the slide)
a3, b3, c3 = 1.5, 0.7, 0.7

# get the result and draw a graph 3
result_3 = odeint(model, state_3, t, args=(Input_3, Enzyme_3, a3, b3, c3))
plot_simulation(t, result_3, 'Simulation 3: I=0.75, x=1, y=1, z=1')
```
