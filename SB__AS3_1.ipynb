```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Initial condition
I = 0.75
X, Y, Z = 1, 1, 1
E_total = 1
time = np.linspace(0, 500, 5000) # 0~500s

# define model
def model(state, time, a, b, c, d, n):
    X, Y, Z = state
    Z = max(0, Z) # Z가 미세한 오차로 음수가 되는 것을 방지 (NaN 에러 차단)
    theta = (Z**n) / (d + Z**n) # define function E = f(z)
    E_active = E_total * theta # decide how much E will be active
    dXdt = I - (a * E_active * X)
    dYdt = (a * E_active * X) - (b * Y)
    dZdt = (b * Y) - (c * Z)
    return [dXdt, dYdt, dZdt]

# define graph function
def graph(time, result):

    # draw a plot
    plt.figure(figsize=(3, 3))
    plt.plot(time, result)

    # set the labels
    plt.xlabel('Time')
    plt.ylabel('Concentrations')

    # set the limits
    plt.xlim(0, 520)
    plt.ylim(0, 5.2)

    # set the ticks
    plt.xticks(ticks=[0, 125, 250, 375, 500], labels=['0', '', '250', '', '500'], fontsize=8)
    plt.yticks(ticks=[0, 1.25, 2.5, 3.75, 5.0], labels=['0', '', '2.5', '', '5.0'], fontsize=8)
    plt.tick_params(axis='both', direction='in')

    # make top, right spines invisible
    ax = plt.gca() # get current axes
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # shows the graph
    plt.show()
```

```python
result = odeint(model, [X, Y, Z], time, args=(1.7, 1, 1, 0.7, 3)) # args = a, b, c, d, n
graph(time, result)
```

```python
result = odeint(model, [X, Y, Z], time, args=(3, 1, 1, 3.3, 2)) # args = a, b, c, d, n
graph(time, result)
```
